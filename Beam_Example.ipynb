# Beam Examples

```python
import Beam
```

```python
v1 = Beam.beam('V1',0.5,0.25,6)

v1.add_pure_bending_moment(4,32)
v1.add_distributed_load([0,6],[-15,-5])
v1.add_point_load(1, -10)
v1.add_point_load(2.5, -15)
v1.add_point_support(0,43.625,2)
v1.add_point_support(4,41.375,1)
```

```python
v1.info_beam()
```

```python
v1.plot_load()
v1.plot_shear()
v1.plot_moment()
```

```python
v2 = Beam.beam('V2',0.5,0.25,6)

v2.add_distributed_load([1,3,5],[0,-4,0])
v2.add_point_load(3, -10)
v2.add_pure_bending_moment(1,30)
v2.add_pure_bending_moment(5,-30)
v2.add_point_support(0,32/6 + 5,2)
v2.add_point_support(6,32/6 + 5,1)
```

```python
v2.info_beam()
```

```python
v2.plot_load()
v2.plot_shear()
v2.plot_moment()
```

```python
v3 = Beam.beam('V3',0.5,0.25,6)

q = -15 # kN/m
l0 = 0 # m
l1 = 6 # m
p = -52 # kN

v3.add_distributed_load([l0,l1],[q,q])
v3.add_point_load((l1-l0)/2, p)
v3.add_point_support(l0,-(q*(l1-l0))/2 + 26,1)
v3.add_point_support(l1,-(q*(l1-l0))/2 + 26,2)
```

```python
v3.info_beam()
```

```python
v3.plot_load()
v3.plot_shear()
v3.plot_moment()
```
